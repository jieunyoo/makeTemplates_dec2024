# finally, process rest of MC samples

```python


sample23= obj[{"Region": "SR1", "Sample": "WQQ"}]
sample24 = obj2[{"Region": "SR1", "Sample": "WQQ"}]
sample = 'WQQ'

for i in getPassList(sample):
    hist1 = sample23[{"Systematic": i}]
    h1 = uproot.to_writable(hist1)
    outFile[f'{sample}_{i}']=h1

for i in getFailList(sample):
    hist1 = sample24[{"Systematic": i}]
    h1 = uproot.to_writable(hist1)
    outFile[f'{sample}_{i}']=h1

sample25= obj[{"Region": "SR1", "Sample": "ZQQ"}]
sample26 = obj2[{"Region": "SR1", "Sample": "ZQQ"}]
sample = 'ZQQ'

for i in getPassList(sample):
    hist1 = sample25[{"Systematic": i}]
    h1 = uproot.to_writable(hist1)
    outFile[f'{sample}_{i}']=h1

for i in getFailList(sample):
    hist1 = sample26[{"Systematic": i}]
    h1 = uproot.to_writable(hist1)
    outFile[f'{sample}_{i}']=h1

```
